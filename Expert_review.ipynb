```python
import pandas as pd 
```

```python

#create a data frame and export data from excel - Jelle
df = pd.read_excel('ExpertReviewsClean43LIWC.xlsx') 
df.head()
```

```python
#indicate which columns I want to keep, and remove the rest - Jelle 
columns_to_keep = ['url', 'idvscore', 'dateP', 'Rev']
df = df[columns_to_keep]
df.head()
```

```python
#give all lines their own key 
df['CR_ID'] = range(1, len(df) +1)
df.head()
```

```python
#changing order of columns where CR_ID is the first column 
desired_order = ['CR_ID', 'idvscore', 'dateP', 'Rev', 'url']
df = df[desired_order]
df.head()
```

```python
#Delete duplicates based in rev and check how many lines are deleted
aantal_duplicaten_voor_verwijderen = df.duplicated(subset=['Rev']).sum()

print(f'Aantal duplicaten te verwijderen: {aantal_duplicaten_voor_verwijderen}')
aantal_rijen = df.shape[0]
print(f"Aantal rijen in de DataFrame: {aantal_rijen}")

df = df.drop_duplicates(subset=['Rev'])
```

```python
df.to_csv('expert_review_clean.csv')
```
